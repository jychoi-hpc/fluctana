```python
A = FluctAna()
A.add_data(KstarEcei(), shot=16150, trange=[7.57, 7.62], clist=['ECEI_G0101-2308'])
A.add_data(KstarEcei(), shot=16150, trange=[7.57, 7.62], clist=['ECEI_G0201-2408'])
```

```python
A.fftbins(nfft=512,window='hann',overlap=0.5,detrend=1)
```

```python
A.coherence(done=0,dtwo=1)
```
