We will plot a few Mirnov coil signals loaded from MDSplus server.

#### Step #0
Import modules.

```python
%matplotlib inline 
from fluctana import *
```

#### Step #1 
Make data set. 

norm = 0 (default = 1) : means no normalization.

res  = 0.000002 : means resampling data with time step = 0.000002 sec.

```python
A = FluctAna()
A.add_data(KstarMds(), shot=18597, trange=[1.8, 1.805], clist=['MC1T06'], norm=0, res=0.000002)
```

#### Step #2
Check channels in the data set.

```python
A.list_data()
```

#### Step #3
Plot channel number 0 in data set number 0 in multiple windows.

```python
A.mplot(dnum=0,cnum=[0],type='time')
```

Plot channel number 0 in data set number 0 in single window (overlapping).

```python
A.oplot(dnum=0,cnum=[0],type='time')
```

Plot spectrums of channel number 0 in data set number 0.

```python
A.spec(dnum=0,cnum=[0],flimits=[0,100000],climits=[-60,-30])
```

Plot (interactively) ECE image of data set number 0 with sample channel number 0

```python
A.iplot(dnum=0,snum=0,vlimits=[-0.1,0.1])
```
