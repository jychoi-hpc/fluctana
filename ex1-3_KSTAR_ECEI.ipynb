```python
A = FluctAna()
A.add_data(KstarEcei(), shot=16150, trange=[7.57, 7.62], clist=['ECEI_G0101-1108', 'ECEI_G1401-2208'])
A.add_data(KstarEcei(), shot=16150, trange=[7.57, 7.62], clist=['ECEI_G0201-1208', 'ECEI_G1501-2308'])
```

```python
A.fftbins(nfft=512,window='hann',overlap=0.5,detrend=1)
```

```python
A.coherence(done=0,dtwo=1)
```

```python
A.cplot(dnum=1,snum=108,frange=[0,60],vlimits=[0,0.4])
```
